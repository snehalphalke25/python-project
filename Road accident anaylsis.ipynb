**ROAD ACCIDENT ANAYLSIS**

**Objective**:
The objective of this analysis is to examine road accident patterns across different states and Union Territories in India during 2018. By analyzing accident data across various dimensions—such as time of day, junction types, severity of injuries, and fatalities—the aim is to identify high-risk areas, periods, and conditions associated with road accidents.

```python
#importing all the libaries
import pandas as pd  # use for data manupulation
import seaborn as sns 
import matplotlib.pyplot as plt # For charts
import numpy as np

# Load the dataset to examine its structure and contents
file_path = "C:/Users/snehal/Downloads/Road Accident Data Set - 2018.csv"
road_accident_data = pd.read_csv(file_path)
print(road_accident_data)


```

```python

import matplotlib.pyplot as plt

# Filter the relevant columns for total accidents by state
states = road_accident_data['States/UTs']
total_accidents = road_accident_data['Total Accidents.1']

# Bar chart
plt.figure(figsize=(12, 8))
plt.bar(states, total_accidents, color='darkblue')
plt.xticks(rotation=90)
plt.xlabel('States/UTs')
plt.ylabel('Total Accidents')
plt.title('Total Accidents by State/UT in 2018')
plt.tight_layout()
plt.show()

```

```python

# Extract relevant columns for injuries at T-Junctions
minor_injury = road_accident_data['T-Junction - Persons Injured - Minor Injury']
grievous_injury = road_accident_data['T-Junction - Persons Injured - Greviously Injured']

# Stacked bar chart
plt.figure(figsize=(12, 8))
plt.bar(states, grievous_injury, label='Grievously Injured', color='orange')
plt.bar(states, minor_injury, bottom=grievous_injury, label='Minor Injury', color='lightgreen')
plt.xticks(rotation=90)
plt.xlabel('States/UTs')
plt.ylabel('Number of Injuries')
plt.title('Injury Severity at T-Junctions by State/UT')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Extract relevant columns for different time periods
time_periods = ['Time 06-0900hrs - (Day)', '09-1200hrs - (Day)', '12-1500hrs - (Day)',
                '15-1800hrs - (Day)', '18-2100hrs - (Night)', '21-2400hrs - (Night)',
                '00-300hrs - (Night)', '03-600hrs - (Night)']
time_data = road_accident_data[time_periods].sum()

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(time_periods, time_data, marker='*', color='purple')
plt.xlabel('Time Period')
plt.ylabel('Total Accidents')
plt.title('Total Accidents by Time of Day')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Sum accidents at T-Junction and Y-Junction across all states
t_junction_accidents = road_accident_data['T-Junction - Total Number of Accidents - Number'].sum()
y_junction_accidents = road_accident_data['Y-Junction - Total Number of Accidents'].sum()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie([t_junction_accidents, y_junction_accidents], labels=['T-Junction', 'Y-Junction'],
        autopct='%1.1f%%', startangle=140, colors=['black', 'grey'])
plt.title('Accident Distribution by Junction Type')
plt.show()

```

```python
# Extract relevant columns for accidents and fatalities at T-Junctions
t_junction_accidents = road_accident_data['T-Junction - Total Number of Accidents - Number']
t_junction_fatalities = road_accident_data['T-Junction - Persons Killed - Number']

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(t_junction_accidents, t_junction_fatalities, color='teal', alpha=0.6)
plt.xlabel('Total Accidents at T-Junction')
plt.ylabel('Fatalities at T-Junction')
plt.title('Scatter Plot of Accidents vs. Fatalities at T-Junctions')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Horizontal bar chart
plt.figure(figsize=(12, 10))
plt.barh(states, grievous_injury, label='Grievously Injured', color='pink')
plt.barh(states, minor_injury, left=grievous_injury, label='Minor Injury', color='red')
plt.xlabel('Number of Injuries')
plt.ylabel('States/UTs')
plt.title('Injury Severity at T-Junctions by State/UT (Horizontal)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Box plot

plt.figure(figsize=(12, 6))
sns.boxplot(data=road_accident_data[time_periods], palette="coolwarm")
plt.title('Distribution of Accidents by Time Period')
plt.xlabel('Time Period')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract and plot accident data for multiple junction types
junction_types = ['T-Junction - Total Number of Accidents - Number', 
                  'Y-Junction - Total Number of Accidents']  # Add more types if available

junction_data = road_accident_data[junction_types].sum()

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(junction_types, junction_data, marker='o', color='darkblue')
plt.xlabel('Junction Type')
plt.ylabel('Total Accidents')
plt.title('Total Accidents Across Different Junction Types')
plt.grid()
plt.tight_layout()
plt.show()
```

CONCLUSION 
Through analyzing road accident data, we identified specific patterns and high-risk areas that contribute to increased accident rates. These patterns reveal that certain times of day, specific intersection types, and particular regions are more prone to accidents.
